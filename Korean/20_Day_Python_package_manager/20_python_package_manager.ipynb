## Python PIP - Python 패키지 관리자

### PIP란?

PIP는 Preferred installer program의 약자입니다. _pip_를 사용하여 다양한 Python 패키지를 설치합니다.
패키지는 하나 이상의 모듈이나 다른 패키지를 포함할 수 있는 Python 모듈입니다. 애플리케이션에 설치할 수 있는 모듈 또는 모듈들이 패키지입니다.
프로그래밍에서 모든 유틸리티 프로그램을 직접 작성할 필요 없이 패키지를 설치하고 애플리케이션에 임포트합니다.

### PIP 설치하기

pip를 설치하지 않았다면 지금 설치해 봅시다. 터미널 또는 명령 프롬프트로 이동하여 다음을 복사하여 붙여넣으세요:

```sh
asabeneh@Asabeneh:~$ pip install pip
```


다음을 입력하여 pip가 설치되었는지 확인하세요

```sh

```

### URL에서 읽기

이제 로컬 머신에 있는 파일을 읽거나 쓰는 방법에 익숙해졌습니다. 때때로 url을 사용하여 웹사이트에서 읽거나 API에서 읽고 싶을 수 있습니다.
API는 Application Program Interface의 약자입니다. 서버 간에 구조화된 데이터, 주로 json 데이터를 교환하는 수단입니다. 네트워크 연결을 열려면 _requests_라는 패키지가 필요합니다. 이 패키지는 네트워크 연결을 열고 CRUD(생성, 읽기, 업데이트, 삭제) 작업을 구현할 수 있게 해줍니다. 이 섹션에서는 CRUD의 읽기 또는 가져오기 부분만 다룰 것입니다.

_requests_를 설치해 봅시다:

````python
asabeneh@Asabeneh:~$ pip install requests
````


_requests_ 모듈에서 _get_, _status_code_, _headers_, _text_ 및 _json_ 메서드를 살펴보겠습니다:
  - _get()_: 네트워크를 열고 url에서 데이터를 가져옵니다 - response 객체를 반환합니다
  - _status_code_: 데이터를 가져온 후 작업 상태를 확인할 수 있습니다 (성공, 오류 등)
  - _headers_: 헤더 타입을 확인합니다
  - _text_: 가져온 response 객체에서 텍스트를 추출합니다
  - _json_: json 데이터를 추출합니다
이 웹사이트에서 txt 파일을 읽어 봅시다, https://www.w3.org/TR/PNG/iso_8859-1.txt.

````python
import requests # request 모듈을 임포트합니다

url = 'https://www.w3.org/TR/PNG/iso_8859-1.txt' # 웹사이트의 텍스트

response = requests.get(url) # 네트워크를 열고 데이터를 가져옵니다
print(response)
print(response.status_code) # 상태 코드, 성공:200
print(response.headers)     # 헤더 정보
print(response.text) # 페이지의 모든 텍스트를 제공합니다
````

### 패키지 만들기

많은 수의 파일을 일정한 기준에 따라 다른 폴더와 하위 폴더로 구성하여 쉽게 찾고 관리할 수 있습니다. 아시다시피 모듈은 클래스, 함수 등 여러 객체를 포함할 수 있습니다. 패키지는 하나 이상의 관련 모듈을 포함할 수 있습니다. 패키지는 실제로 하나 이상의 모듈 파일을 포함하는 폴더입니다. 다음 단계를 사용하여 mypackage라는 패키지를 만들어 봅시다:

30DaysOfPython 폴더 안에 mypackage라는 새 폴더를 만드세요
mypackage 폴더에 빈 **__init__**.py 파일을 만드세요.
다음 코드로 arithmetic.py 및 greet.py 모듈을 만드세요:

````python
# mypackage/arithmetics.py
# arithmetics.py
def add_numbers(*args):
    total = 0
    for num in args:
        total += num
    return total


def subtract(a, b):
    return (a - b)


def multiple(a, b):
    return a * b


def division(a, b):
    return a / b


def remainder(a, b):
    return a % b


def power(a, b):
    return a ** b
````

````python
# mypackage/greet.py
# greet.py
def greet_person(firstname, lastname):
    return f'{firstname} {lastname}, welcome to 30DaysOfPython Challenge!'
````


패키지의 폴더 구조는 다음과 같아야 합니다:

```sh
─ mypackage
    ├── __init__.py
    ├── arithmetic.py
    └── greet.py
```

보시다시피 패키지가 완벽하게 작동합니다. 패키지 폴더에는 **__init__**.py라는 특별한 파일이 있으며 패키지의 내용을 저장합니다. 패키지 폴더에 **__init__**.py를 넣으면 Python이 이를 패키지로 인식하기 시작합니다.
**__init__**.py는 해당 모듈의 지정된 리소스를 다른 Python 파일에서 임포트할 수 있도록 노출합니다. 빈 **__init__**.py 파일은 패키지가 임포트될 때 모든 함수를 사용 가능하게 만듭니다. **__init__**.py는 폴더가 Python에서 패키지로 인식되는 데 필수적입니다.

  - Numpy: Numpy(numeric python)는 Python에서 가장 인기 있는 머신 러닝 라이브러리 중 하나로 알려져 있습니다.
  - Pandas: Python에서 고수준 데이터 구조와 다양한 분석 도구를 제공하는 데이터 분석, 데이터 과학 및 머신 러닝 라이브러리입니다.
  - SciPy: SciPy는 애플리케이션 개발자와 엔지니어를 위한 머신 러닝 라이브러리입니다. SciPy 라이브러리에는 최적화, 선형 대수, 적분, 이미지 처리 및 통계 모듈이 포함되어 있습니다.
  - Scikit-Learn: NumPy와 SciPy 기반입니다. 복잡한 데이터를 처리하기 위한 최고의 라이브러리 중 하나로 간주됩니다.
  - TensorFlow: Google에서 만든 머신 러닝 라이브러리입니다.
  - Keras: Python에서 가장 멋진 머신 러닝 라이브러리 중 하나로 간주됩니다. 신경망을 표현하기 위한 더 쉬운 메커니즘을 제공합니다. Keras는 또한 모델 컴파일, 데이터셋 처리, 그래프 시각화 등을 위한 최고의 유틸리티를 제공합니다.
- 네트워크:
  - requests: 서버에 요청(GET, POST, DELETE, PUT)을 보내는 데 사용할 수 있는 패키지입니다
    - _pip install requests_
