### Arithmetische Operatoren:

- Addition(+): a + b
- Subtraktion(-): a - b
- Multiplikation(*): a * b
- Division(/): a / b
- Modulus(%): a % b
- Ganzzahldivision(//): a // b
- Potenzierung(**): a ** b

![Arithmetische Operatoren](../images/arithmetic_operators.png)

**Beispiel: Ganzzahlen**

```py

print('Division ohne Rest: ', 7 // 2)   # 3, ergibt keine Gleitkommazahl oder keinen Rest
print ('Division ohne Rest: ',7 // 3)   # 2
print('Modulus: ', 3 % 2)         # 1, Gibt den Rest zur√ºck
print('Potenzierung: ', 2 ** 3) # 8 es bedeutet 2 * 2 * 2
```

**Beispiel: Gleitkommazahlen**

In der Programmierung vergleichen wir Werte, wir verwenden Vergleichsoperatoren, um zwei Werte zu vergleichen. Wir pr√ºfen, ob ein Wert gr√∂√üer oder kleiner oder gleich einem anderen Wert ist. Zus√§tzlich zu den oben genannten Vergleichsoperatoren verwendet Python:

### Logische Operatoren

Im Gegensatz zu anderen Programmiersprachen verwendet Python die Schl√ºsselw√∂rter _and_, _or_ und _not_ f√ºr logische Operatoren. Logische Operatoren werden verwendet, um bedingte Anweisungen zu kombinieren:

![Logische Operatoren](../images/logical_operators.png)

```py

print('True and True: ', True and True)
print(3 > 2 or 4 > 3)  # True - weil beide Aussagen wahr sind
print(3 > 2 or 4 < 3)  # True - weil eine der Aussagen wahr ist
print(3 < 2 or 4 < 3)  # False - weil beide Aussagen falsch sind
print('True or False:', True or False)
print(not 3 > 2)     # False - weil 3 > 2 wahr ist, dann ergibt not True False
print(not True)      # False - Negation, der not-Operator verwandelt True in False
print(not False)     # True
print(not not True)  # True
print(not not False) # False

```
